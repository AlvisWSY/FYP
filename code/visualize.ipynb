```python
import pandas as pd
import matplotlib.pyplot as plt
# 读取CSV文件
path2='/media/mldadmin/home/s123mdg34_04/WangShengyuan/FYP/data/internal/cleaned/corona/2022-12-02_12-37-38_ch0-WaveForm_G2-L256-T2120-Prt32_corona.csv'
df = pd.read_csv(path2)
random_rows = df.sample(n=5)

# 对于每一行（即每一个波形），绘制图形
for index, row in random_rows.iterrows():
    plt.figure()  # 创建新的图形
    plt.plot(row)  # 绘制波形
    plt.title(f'Waveform {index}')  # 设置标题
    plt.show()  # 显示图形
```
