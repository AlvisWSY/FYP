```python
import pandas as pd
import matplotlib.pyplot as plt
# read the csv file
path1='/media/mldadmin/home/s123mdg34_04/WangShengyuan/FYP/data/corona/cleaned/corona/2022-12-02_16-57-31_ch0-WaveForm_G10-L256-T1519-Prt32_corona.csv'
df = pd.read_csv(path1)

# randomly select 5 rows
random_rows = df.sample(n=5)

# plot the waveforms
for index, row in random_rows.iterrows():
    plt.figure()  
    plt.plot(row)  
    plt.title(f'Waveform {index}')  
    plt.show()  
```
