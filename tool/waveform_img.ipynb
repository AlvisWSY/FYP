```python
import csv
import matplotlib.pyplot as plt

# Define the path to the CSV file
Path = '/media/mldadmin/home/s123mdg34_04/WangShengyuan/FYP/data/corona/waveform_CSV/2022-12-02_16-57-31_ch0-WaveForm_G10-L256-T1519-Prt32.csv'

# Read the CSV file
with open(Path, 'r') as file:
    reader = csv.reader(file, delimiter='\t')
    next(reader)  # Skip the header row
    waveform_data = next(reader)  # Get the second row of waveform data

# Convert the waveform data to a list of floats
waveform = [float(value) for value in waveform_data]

# Plot the waveform
plt.plot(waveform)
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.title('Waveform Graph')
plt.show()
```
